```python
import pandas as pd
import numpy as np


n = pd.read_excel("Python_Test.xlsm") # 讀檔
# 轉成 dataframe 後刪掉不要的欄位
df = pd.DataFrame(n)
df = df.drop("Q1: 右表有四百多隻基金，請用python寫程式，去公開網站爬蟲下載最新的基金月報",axis = 1)
df = df.drop("Unnamed: 1",axis = 1)
name_list = list(n["c_name"])
for (i,item) in enumerate(name_list):
    if "證券投資信託" in item: 
        name_list[i] = item.replace("證券投資信託","")
df
```

```python
import pandas as pd
import pdfplumber
import re

# def function#
########################################################################################
def source_result(raw_list):
    source_label = ["lipper", "理柏", "morningstar", "晨星", "其他"]
    result =[]
    for item in source_label:
        for item2 in raw_list:
            if item in item2:
                result.append(item)
            else:
                result.append("其他")
    return set(result)     # 用 set 去除重複值後回傳

#########################################################################################
error = [] 

for (ind,fund) in enumerate(name_list):
    try:
        pdf = pdfplumber.open( fund + ".pdf")
        source = ""
        seq = ""
        for (p,page) in enumerate(pdf.pages): # 頁數是list, extract_text()是str
            for(t,text) in enumerate(page.extract_text()):
                seq = seq + text  # 取出來是一個一個字，串成字串
        find_word = "資料來源"  
        word_ind = [m.start() for m in re.finditer(find_word, seq)] # 找"資料來源"在月報中的位置
        for l in word_ind:
            for num in range(l+5, l+15):
                source = source + seq[num].lower() # 英文全部轉小寫, 中文會 ignore
                if seq[num] == " " and seq[num-1]!=":" : break
        source_list = source.split()
        if "柏瑞" in fund:                # 柏瑞資產的報告中「資料來源」這幾個字不是文字，所以讀不到
            df.iloc[ind][2] = set(list(lipper))    # 但經觀察，柏瑞資料來源都只有 lipper，所以另外自己補上
        else:
            df.iloc[ind][2] = source_result(source_list)
    except:
        error.append(fund)
df  # 資料來源為問號者，代表爬蟲下載月報失敗，所以沒有檔按可以看資料來源
```
