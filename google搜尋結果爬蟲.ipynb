```python
import pandas as pd
import numpy as np


n = pd.read_excel("Python_Test.xlsm")
name_list = list(n["c_name"])
for (i,item) in enumerate(name_list):
    if "證券投資信託" in item: 
        name_list[i] = item.replace("證券投資信託","")
     
pd.DataFrame(name_list) # 取得基金列表
```

```python
from bs4 import BeautifulSoup
from urllib.parse import quote
import time
import requests
from selenium import webdriver
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.chrome.options import Options

errors = []
for (ind,fund) in enumerate(name_list):
    print(ind)
    try:
        url = "https://www.google.com/search?q=" + "鉅亨"+ fund + "月報"  # 在google搜尋基金名稱
        res = requests.get(url)
        soup = BeautifulSoup(res.text,"html.parser")

        for item in soup.find_all("a"):          # 取得第一個搜尋結果
            if "cnyes" in item.get("href"):
                s = str(item) 
                end = s.index("><")              # 從網頁原始碼中找出基金網頁網址
                begin = s.index("https")
                search_result_url = s[begin:end-1]
                break
                        
        res = requests.get(search_result_url) 
        soup = BeautifulSoup(res.text,"html.parser")
        for link in soup.find_all("a", href = True):
                s = str(link)
                if "報告下載" in s: 
                    end = s.index(">")        # 取得基金月報網頁網址
                    begin = s.index("detail")
                    report_url = "https://fund.cnyes.com/" + s[begin:end-1]
                    break
        
        driver = webdriver.Ie() # 須將 IE 中保護選項關閉，chrome剛出新版，selenium尚不能support新版 chrom
        driver.get(report_url)
        xpath = "//*[@id=" + '"'+"anue-ga-wrapper" +'"' +"]/div/div[2]/div[2]/main[2]/section/div[2]/a[1]"
        report = driver.find_element_by_xpath(xpath).get_attribute('href') # 報告
        r = requests.get(report) 
        with open(fund +".pdf",'wb') as f:   # 將報告存檔
            f.write(r.content)
            driver.close()
        time.sleep(5)   # 速度太快會被網頁封鎖，所以每一筆都停一下在繼續做 (原本用3秒結果不夠長)
    except: 
        errors.append(fund) # 下載失敗的基金名稱列表
        print("error")
        time.sleep(5)
```

```python
total = len(name_list)
success = len(name_list) - len(errors)
fail = len(errors)
print("總筆數: ", total)
print("成功筆數: ", success)
print("失敗筆數: ", fail)
print("成功率: ", round(success/total, 2))
```
